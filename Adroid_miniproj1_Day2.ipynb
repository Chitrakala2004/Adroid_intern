```python
from itertools import combinations

set_num = {-12, -3, -6, 7, 2, -2, 6, 3, 9, -7, -5, -8, 1, 11, -9, -4}

subset_size = 5

def find_zero_sum_subsets(set_num, subset_size):
    zero_sum_subsets = []
    for subset in combinations(set_num, subset_size):
        if sum(subset) == 0:
            zero_sum_subsets.append(subset)
    return zero_sum_subsets

zero_sum_subsets = find_zero_sum_subsets(set_num, subset_size)
for subset in zero_sum_subsets:
    print(subset)
```

```python
import random
from itertools import combinations

numbers = {-12, -3, -6, 7, 2, -2, 6, 3, 9, -7, -5, -8, 1, 11, -9, -4}

min_size = 3
max_size = 6

max_attempts = 10000

def find_zero_sum_subsets_random(numbers, min_size, max_size, max_attempts):
    zero_sum_subsets = set()
    attempts = 0
    while attempts < max_attempts:
        subset_size = random.randint(min_size, max_size)
        subset = tuple(random.sample(numbers, subset_size))
        if sum(subset) == 0:
            zero_sum_subsets.add(subset)
        attempts += 1
    return zero_sum_subsets

zero_sum_subsets = find_zero_sum_subsets_random(numbers, min_size, max_size, max_attempts)
for subset in zero_sum_subsets:
    print(subset)
```
